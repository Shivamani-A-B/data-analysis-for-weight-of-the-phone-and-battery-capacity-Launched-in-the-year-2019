```python
#taking mode for categorial values to fill the  null values.
df['Removable storage'].fillna(str(df['Removable storage'].mode().values[0]),inplace=True)
df['Custom launcher'].fillna(str(df['Custom launcher'].mode().values[0]),inplace=True)
df['Facial recognition'].fillna(str(df['Facial recognition'].mode().values[0]),inplace=True)
```

```python
df.isnull().sum()
```

```python
#replacing the symbols and signs into numbers
df['Battery']=df['Battery'].apply(lambda x: str(x).replace('mAh','')if 'mAh' in str(x) else str(x))
df['Battery']=df['Battery'].apply(lambda x: int(x))
```

```python
#grouping the values based on model
grp=df.groupby('Model')
x=grp['Battery'].agg(np.mean)
y=grp['Weight in g'].agg(np.mean)
print(x)
print(y)
```

# Plotting the data for analysis

```python
#plotting the model in terms of battery for analysis
plt.figure(figsize=(35,10))
plt.plot(x,color='b')
plt.xticks(rotation=90)
plt.title('Model wise visualisation with battery capacity')
plt.xlabel("Models")
plt.ylabel("Battery in mAh")
plt.show()
```

```python
# plotting the model in terms of weight in grams.
plt.figure(figsize=(35,10))
plt.plot(y,"ro",color='g')
plt.xticks(rotation=90)
plt.title('Model wise visualisation with Weights in grams')
plt.xlabel("Models")
plt.ylabel("Weight in grams")
plt.show()
```
